```python
#Nashville Housing Data - what is the best season for a single family to to sell a property?
import pandas as pd
```

```python
#Unzip archive to Jupiter directory
path_to_zip_file = "nashville-housing-data.zip"
directory_to_extract_to = ""
import zipfile
zip_ref = zipfile.ZipFile(path_to_zip_file, 'r')
zip_ref.extractall(directory_to_extract_to)
zip_ref.close()
```

```python
#Reading data in a csv file
df = pd.read_csv('Nashville_housing_data_2013_2016.csv')
df
```

```python
df.describe()
```

```python
#Dropping 2 first columns as they don't have relevant data
df = df.drop('Unnamed: 0', axis=1)
df = df.drop("Unnamed: 0.1", axis=1)
df.head()
```

```python
#Creating function to extract month from Sales Date
#vals=1 as Month is on second place in data 2013-01-24 (year = 0, month=1, date=2). 
#If we need to get year of sale, vals equals to 0, etc.
import numpy as np

def getMonth(x) :
    vals = x.split("-")
    return vals[1]
df['Sale Month'] = df['Sale Date'].apply(getMonth)
df.head()
```

```python
#Creating function to convert Month into a season
def getSeason(x):
    if x=="12":
        return "Winter"
    if x=="01":
        return "Winter"
    if x=="02":
        return "Winter"
    if x=="03":
        return "Spring"
    if x=="04":
        return "Spring"
    if x=="05":
        return "Spring"
    if x=="06":
        return "Summer"
    if x=="07":
        return "Summer"
    if x=="08":
        return "Summer"
    if x=="09":
        return "Fall"
    if x=="10":
        return "Fall"
    if x=="11":
        return "Fall"
df['Sale Season'] = df['Sale Month'].apply(getSeason)
df.head()
```
